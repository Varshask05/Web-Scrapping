```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://www.freelancer.in/jobs"
page = requests.get(URL)
```

```python
soup = BeautifulSoup(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
job_listings = soup.find_all('div', class_='JobSearchCard-item')
job_listings
```

```python
# Extract data from each job listing
for job in job_listings:
    job_title = job.find('a', class_='JobSearchCard-primary-heading-link').text.strip()
    job_description = job.find('p', class_='JobSearchCard-primary-description').text.strip()
    job_requirements = job.find('p', class_='JobSearchCard-primary-skills').text.strip()
    bid_price = job.find('span', class_='JobSearchCard-primary-price').text.strip()

    print("Job Title:", job_title)
    print("Description:", job_description)
    print("Requirements:", job_requirements)
    print("Bid Price:", bid_price)
    print("=" * 50)
```
